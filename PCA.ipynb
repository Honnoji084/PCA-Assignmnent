# HIERARCHICAL CLUSTERING

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
#Create dendogram of normalized data for all methods
for methods in ['single','complete','average','weighted','centroid','median','ward']:
    plt.figure(figsize=(25,8))
    plt.title('Visualising the data Method - {}'.format(methods))
    Dendrogram = sch.dendrogram(sch.linkage(wine_norm,method = methods,optimal_ordering=False))
```

```python
from sklearn.metrics import silhouette_score as sil
nc=[2,3,4,5,6,7,8,9]
sc_score=[]
for i in nc:
    hc=AgglomerativeClustering(n_clusters=i,affinity="euclidean",linkage="ward")
    
    y=hc.fit_predict(wine_norm)
    sc_score.append(sil(wine_norm,y))
```

```python
plt.plot(nc,sc_score,color="b")
plt.scatter(nc,sc_score,color="r")
plt.show()
```

### By seeing dendrogram and silhouette score the best number of cluster we are getting is 3 and linkage method is ward

# K-MEANS 

```python
from sklearn.cluster import KMeans
```

```python
nc=[2,3,4,5,6,7,8,9]
wcss=[]
for i in nc:
    km=KMeans(i)
    km.fit(wine_norm)
    wcss.append(km.inertia_)
```

```python
plt.plot(nc,wcss,color='b')
plt.title('ELBOW METHOD')
plt.xlabel('No. Of Clusters')
plt.ylabel('WCSS')
plt.scatter(nc,wcss,color='r')
plt.show()
```

```python
# model Building
model_kmeans=KMeans(n_clusters=3)
model_kmeans.fit(wine_norm)
```

```python
wine3=wine
```

```python
wine3['Norm_KMClusters']=pd.Series(model_kmeans.labels_)
```
